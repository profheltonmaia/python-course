## Manipulação de Arrays com o NumPy

[NumPy](https://numpy.org/) é uma biblioteca fundamental para computação numérica em Python. Ela fornece estruturas de dados eficientes para a manipulação de arrays multidimensionais, bem como funções matemáticas para operações rápidas em elementos desses arrays.

A instrução `import numpy as np` é utilizada para importar a biblioteca NumPy e atribuir a ela um nome de apelido (alias), que neste caso é "np". Essa prática é comum e recomendada quando se trabalha com bibliotecas extensas como o NumPy, pois facilita a referência a funções e objetos da biblioteca, tornando o código mais conciso e legível.

### Definindo Arrays

```python
# Importando a biblioteca NumPy
import numpy as np

# Criando uma array unidimensional
array_1d = np.array([1, 2, 3, 4, 5])

# Criando uma array bidimensional
array_2d = np.array([[1, 2, 3], [4, 5, 6]])

# Arrays especiais
zeros_array = np.zeros((3, 3))  # Array 3x3 preenchido com zeros
ones_array = np.ones((2, 4))   # Array 2x4 preenchido com uns
random_array = np.random.rand(3, 2)  # Array 3x2 com valores aleatórios entre 0 e 1

# Imprimindo os arrays
print("Array Unidimensional:",array_1d)
print("Array Bidimensional:",array_2d)
print("Array Zeros:",zeros_array)
print("Array Ones:",ones_array)
print("Array Random:",random_array)
```

## Visualização de Dados com o Matplotlib

**Visualização de Dados com [Matplotlib](https://matplotlib.org/)**

A visualização de dados é essencial para analisar e compreender informações, proporcionando clareza na apresentação de dados e facilitando a identificação de padrões e tendências.

O Matplotlib, uma biblioteca de visualização de dados em Python, é uma escolha popular entre cientistas de dados, engenheiros e analistas de negócios. Além de ser de código aberto e gratuito, destaca-se por sua facilidade de uso.

**Funcionalidades do Matplotlib**

A biblioteca oferece uma ampla gama de funcionalidades, incluindo gráficos de linha, barras, dispersão, área, pizza, histogramas, boxplots, violin plots, mapas de calor, nuvem de palavras, entre outros. A personalização de gráficos também é uma característica marcante.

**Exemplos de Visualizações com Matplotlib**

- Gráfico de linha: Evolução de casos de COVID-19 no Brasil.
- Gráfico de barras: Distribuição de renda no Brasil.
- Gráfico de dispersão: Correlação entre duas variáveis.
- Histograma: Distribuição de uma variável.

**Integração com Outras Bibliotecas**

O Matplotlib integra-se de maneira eficiente com outras bibliotecas, como NumPy, proporcionando um ambiente poderoso para análise de dados. Essa integração permite, por exemplo, visualizar dados gerados por algoritmos de aprendizado de máquina.

### Gráficos Simples
Para introduzir a criação de gráficos, começaremos com um exemplo simples de gráfico de linha usando a biblioteca Matplotlib.

````python
import matplotlib.pyplot as plt

# Dados de exemplo
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Criando o gráfico de linha
plt.plot(x, y, marker='o', linestyle='-', color='b', label='Dados de Exemplo')

# Adicionando rótulos e título
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Gráfico de Linha Simples')
plt.legend()

# Exibindo o gráfico
plt.show()
````

O código começa importando a biblioteca Matplotlib usando `import matplotlib.pyplot as plt`. Isso é essencial para utilizar as funcionalidades da biblioteca, e o alias `plt` é comumente usado para facilitar o acesso a essas funções. Em seguida, o código continua com a definição dos dados e a criação do gráfico de linha. Ao final, a função `show()` é utilizada para exibir o gráfico.

### Gráfico de Linha: Evolução de Casos de COVID-19 no Brasil

Este exemplo utiliza o Matplotlib para criar um gráfico de linha representando a evolução dos casos de COVID-19 no Brasil. A biblioteca NumPy é utilizada para gerar dados fictícios para ilustração.

````python
import matplotlib.pyplot as plt
import numpy as np

# Dados fictícios para ilustração
dias = np.arange(1, 31)
casos_diarios = np.random.randint(100, 1000, size=30).cumsum()

# Criando o gráfico de linha
plt.plot(dias, casos_diarios, marker='o', linestyle='-', color='b', label='Casos Diários')

# Adicionando rótulos e título
plt.xlabel('Dias')
plt.ylabel('Total de Casos')
plt.title('Evolução de Casos de COVID-19 no Brasil')
plt.legend()

# Exibindo o gráfico
plt.show()
````

### Gráfico de Barras: Distribuição de Renda no Brasil
Neste exemplo, um gráfico de barras é utilizado para representar a distribuição de renda no Brasil. Os dados são fictícios e representam percentuais da população em diferentes faixas de renda.

### Gráfico de Dispersão: Correlação entre Duas Variáveis

Este exemplo utiliza um gráfico de dispersão para visualizar a correlação entre duas variáveis fictícias: horas de estudo e desempenho em exames.

````python
import matplotlib.pyplot as plt
import numpy as np

# Dados fictícios para ilustração
horas_estudo = np.random.uniform(1, 10, size=50)
desempenho_exames = 70 + 2 * horas_estudo + np.random.normal(0, 2, size=50)

# Criando o gráfico de dispersão
plt.scatter(horas_estudo, desempenho_exames, color='red', alpha=0.7)

# Adicionando rótulos e título
plt.xlabel('Horas de Estudo')
plt.ylabel('Desempenho em Exames')
plt.title('Correlação entre Horas de Estudo e Desempenho em Exames')

# Exibindo o gráfico
plt.show()
````

### Histograma: Distribuição de uma Variável

Neste exemplo, um histograma é utilizado para representar a distribuição de idades em uma amostra de pessoas.


````python
import matplotlib.pyplot as plt
import numpy as np

# Dados fictícios para ilustração
idades = np.random.normal(30, 5, size=1000)

# Criando o histograma
plt.hist(idades, bins=20, color='purple', edgecolor='black')

# Adicionando rótulos e título
plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.title('Distribuição de Idades')

# Exibindo o gráfico
plt.show()
````

### Explorando Gráficos Estatísticos

Neste exemplo simplificado, abordaremos a representação gráfica de uma distribuição gaussiana bidimensional, utilizando subplots para destacar diferentes características.

Começamos definindo uma distribuição gaussiana bidimensional com média zero e uma matriz de covariância específica. A primeira subplot exibe a densidade de probabilidade, proporcionando uma visão visual da dispersão dos dados nos eixos X e Y.

As subplots seguintes exploram as projeções marginais, revelando a densidade de probabilidade ao longo do eixo X e Y, respectivamente.


````python
import matplotlib.pyplot as plt
import numpy as np

# Definindo os parâmetros da distribuição gaussiana
mean = [0, 0]
covariance = [[1, 0.5], [0.5, 1]]

# Criando uma grade de valores para x e y
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
x, y = np.meshgrid(x, y)

# Calculando os valores da distribuição gaussiana manualmente
inv_covariance = np.linalg.inv(covariance)
det_covariance = np.linalg.det(covariance)
exponent = -0.5 * (x**2 * inv_covariance[0, 0] + y**2 * inv_covariance[1, 1] + 2 * x * y * inv_covariance[0, 1])
z = (1 / (2 * np.pi * np.sqrt(det_covariance))) * np.exp(exponent)

# Criando subplots
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# Subplot da densidade de probabilidade
axs[0].contourf(x, y, z, cmap='viridis')
axs[0].set_title('Densidade de Probabilidade')
axs[0].set_xlabel('Eixo X')
axs[0].set_ylabel('Eixo Y')

# Projeção bidimensional no eixo X
axs[1].plot(x[0, :], np.sum(z, axis=0), color='blue')
axs[1].set_title('Projeção no Eixo X')
axs[1].set_xlabel('Eixo X')
axs[1].set_ylabel('Densidade de Probabilidade')

# Projeção bidimensional no eixo Y
axs[2].plot(np.sum(z, axis=1), y[:, 0], color='red')
axs[2].set_title('Projeção no Eixo Y')
axs[2].set_xlabel('Densidade de Probabilidade')
axs[2].set_ylabel('Eixo Y')

# Ajustando o layout
plt.tight_layout()

# Exibindo os subplots
plt.show()
````

Esse código possui muitos detalhes matemáticos e de configuração, vamos analisar passo a passo:

1. **Definição dos Parâmetros da Distribuição Gaussiana:**
   - `mean = [0, 0]`: Define a média da distribuição gaussiana bidimensional como (0, 0).
   - `covariance = [[1, 0.5], [0.5, 1]]`: Define a matriz de covariância da distribuição gaussiana bidimensional.

2. **Criação da Grade de Valores para x e y:**
   - `x = np.linspace(-3, 3, 100)`: Cria uma sequência linear de 100 valores entre -3 e 3 para o eixo x.
   - `y = np.linspace(-3, 3, 100)`: Cria uma sequência linear de 100 valores entre -3 e 3 para o eixo y.
   - `x, y = np.meshgrid(x, y)`: Cria uma grade bidimensional de valores combinando os valores de x e y.

3. **Cálculo dos Valores da Distribuição Gaussiana Manualmente:**
   - `inv_covariance = np.linalg.inv(covariance)`: Calcula a matriz inversa da matriz de covariância.
   - `det_covariance = np.linalg.det(covariance)`: Calcula o determinante da matriz de covariância.
   - `exponent = -0.5 * (x**2 * inv_covariance[0, 0] + y**2 * inv_covariance[1, 1] + 2 * x * y * inv_covariance[0, 1])`: Calcula o expoente da função gaussiana.
   - `z = (1 / (2 * np.pi * np.sqrt(det_covariance))) * np.exp(exponent)`: Calcula os valores da distribuição gaussiana bidimensional.

4. **Criação de Subplots:**
   - `fig, axs = plt.subplots(1, 3, figsize=(15, 5))`: Cria uma figura com uma linha e três colunas de subplots.

5. **Subplot da Densidade de Probabilidade:**
   - `axs[0].contourf(x, y, z, cmap='viridis')`: Plota a densidade de probabilidade utilizando contornos preenchidos.
   - Configuração de título e rótulos dos eixos.

6. **Projeção Bidimensional no Eixo X:**
   - `axs[1].plot(x[0, :], np.sum(z, axis=0), color='blue')`: Plota a projeção da densidade de probabilidade no eixo X.
   - Configuração de título e rótulos dos eixos.

7. **Projeção Bidimensional no Eixo Y:**
   - `axs[2].plot(np.sum(z, axis=1), y[:, 0], color='red')`: Plota a projeção da densidade de probabilidade no eixo Y.
   - Configuração de título e rótulos dos eixos.

8. **Ajuste de Layout e Exibição dos Subplots:**
   - `plt.tight_layout()`: Ajusta automaticamente o layout dos subplots para evitar sobreposição.
   - `plt.show()`: Exibe os subplots.

Este código cria uma representação visual completa de uma distribuição gaussiana bidimensional, oferecendo diferentes perspectivas por meio de subplots.
